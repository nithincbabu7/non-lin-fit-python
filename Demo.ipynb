```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

from analysis import NonLinearFit
```

```python
score1 = np.linspace(0, 1, 1000)
score2 = 100*(1-score1**3) + 5*np.random.randn(1000)
# score2 = np.clip(100*(score1**3) + 2*np.random.randn(1000), 0, 100)

matplotlib.rcParams['figure.figsize'] = [8, 8]
plt.scatter(score2, score1, s=2)
```

```python
nlinfit = NonLinearFit(score1, score2)
fitted_score = nlinfit.transform(score2)

x_lin = np.linspace(np.min(score2), np.max(score2), 1000)
y_lin = nlinfit.transform(x_lin)

matplotlib.rcParams['figure.figsize'] = [16, 8]
plt.subplot(1, 2, 1)
plt.scatter(score2, score1, s=2)
plt.plot(x_lin, y_lin, c='red')
plt.xlabel('Prediction')
plt.ylabel('MOS')

plt.subplot(1, 2, 2)
plt.scatter(fitted_score, score1, s=2)
plt.xlabel('Fitted Prediction')
plt.ylabel('MOS')
plt.plot([np.min(fitted_score), np.max(fitted_score)], [np.min(fitted_score), np.max(fitted_score)], c='red')
```

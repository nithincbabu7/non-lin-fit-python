```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

from analysis import NonLinearFit
```

```python
# generate non-linear data (replace this with the data used for analysis)
score1 = np.linspace(0, 1, 1000)
score2 = 100*(1-score1**3) + 5*np.random.randn(1000)

matplotlib.rcParams['figure.figsize'] = [8, 8]
plt.scatter(score2, score1, s=2)
plt.title('Scatter Plot')
plt.xlabel('score1')
plt.ylabel('score2')
```

```python
# find a non-linear fit to map score1 to score2
nlinfit = NonLinearFit(score1, score2)

# use the learnt fitting parameters to trensform score1
fitted_score = nlinfit.transform(score2)

# the lines below are only for illustration purpose
x_lin = np.linspace(np.min(score2), np.max(score2), 1000)
y_lin = nlinfit.transform(x_lin)

matplotlib.rcParams['figure.figsize'] = [16, 8]
plt.subplot(1, 2, 1)
plt.scatter(score2, score1, s=2)
plt.plot(x_lin, y_lin, c='red')
plt.title('Scatter Plot with fitting curve')
plt.xlabel('score1')
plt.ylabel('score2')

plt.subplot(1, 2, 2)
plt.scatter(fitted_score, score1, s=2)
plt.title('Scatter Plot after non_linear fitting')
plt.xlabel('fitted_score1')
plt.ylabel('score2')
plt.plot([np.min(fitted_score), np.max(fitted_score)], [np.min(fitted_score), np.max(fitted_score)], c='red')
```
